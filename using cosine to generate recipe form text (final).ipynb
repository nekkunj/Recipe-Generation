### using cosine similairyt to generate recipe id

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from spellchecker import SpellChecker


spell = SpellChecker()

# Input text
input_text = "spinach pasta "

# Function to preprocess text
def preprocess(text):
    text=str(text)
    return ' '.join(text.lower().replace('.', '').split())
    
    
def preprocess2(text):
    text=str(text)
    corrected_words = [spell.correction(word) for word in text.lower().replace('.', '').split()]
    return ' '.join(corrected_words)

# Preprocess the input text and keywords
input_text_clean = preprocess2(input_text)
df['Keywords_clean'] = df['extracted_keywords'].apply(preprocess)

# Calculate similarity using TF-IDF and Cosine Similarity
vectorizer = TfidfVectorizer()
all_texts = df['Keywords_clean'].tolist() + [input_text_clean]
tfidf_matrix = vectorizer.fit_transform(all_texts)
cosine_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])

# Find the best match
best_idx = cosine_sim.argmax()
best_match = df.iloc[best_idx]

print(f"Best Matched Row: {best_match['RecipeId']} ")
```

### using LSTM to generate recipe id

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Sample data
texts = ["I like apples and oranges", "Fruit market is near"]
keywords = ["apple orange", "banana mango", "grape apple", "orange banana grape"]

# Tokenization and sequence padding
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts + keywords)
sequences = tokenizer.texts_to_sequences(texts + keywords)
padded_sequences = pad_sequences(sequences, padding='post')

# Model setup
model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=50, input_length=padded_sequences.shape[1]),
    LSTM(64),
    Dense(len(keywords), activation='sigmoid')  # One output unit per keyword set
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Example data preparation for training (this is a dummy setup)
# Normally, you would have some target labels indicating the match between texts and keywords
targets = np.array([[1, 0, 1, 0], [0, 1, 0, 1]])  # Example labels

# Training
model.fit(padded_sequences[:len(texts)], targets, epochs=10)

# Inference
predictions = model.predict(padded_sequences[len(texts):])
print(predictions)
```
